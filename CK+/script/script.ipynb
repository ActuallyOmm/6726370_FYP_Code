CK+ Dataset
All images are stored in a csv file
* Images have an emotion class no.
* Images have a string of pixel values
* Images have a train/val/test tag

This script is responsible for translating the pixels values in the script and seperating it into the relevant class and emotions

```python
import pandas as pd
import numpy as np
import numpy as np
import pandas as pd
import os
from PIL import Image
```

```python
def convertStringToNumPyArray(pixelString):
    pixel_values = np.array([int(x) for x in pixelString.split()])
    side_length = int(np.sqrt(len(pixel_values)))
    image = pixel_values.reshape((side_length, side_length))
    return image

os.chdir('/user/HS401/ob00564/Documents/COM3001/CK+/script')
ckDataFrame = pd.read_csv("ckextended.csv")
# Reference of how to apply transform to DataFrame
# Resource: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html
ckDataFrame["pixels"] = ckDataFrame["pixels"].apply(convertStringToNumPyArray)
print(type(ckDataFrame["pixels"]))
```

```python
# Create dataframes for each data set, removing contempt as an emotion

# How to filter dataframe
# Resource: https://www.geeksforgeeks.org/ways-to-filter-pandas-dataframe-by-column-values/
train_df = ckDataFrame[ckDataFrame['Usage'] == 'Training']
train_df = train_df[train_df['emotion'] != 7]

val_df = ckDataFrame[ckDataFrame['Usage'] == 'PublicTest']
val_df = val_df[val_df['emotion'] != 7]

test_df = ckDataFrame[ckDataFrame['Usage'] == 'PrivateTest']
test_df = test_df[test_df['emotion'] != 7]
```

```python
# Current wokring directory is within CK+ folder
# Creating new directories for each dataframe
# Resource: https://www.geeksforgeeks.org/create-a-directory-in-python/
os.getcwd()
def createDir(directory_name):
    try:
        os.mkdir(directory_name)
        print(f"Directory '{directory_name}' created successfully.")
    except FileExistsError:
        print(f"Directory '{directory_name}' already exists.")
    except PermissionError:
        print(f"Permission denied: Unable to create '{directory_name}'.")
    except Exception as e:
        print(f"An error occurred: {e}")

def createSubDir(directory_name):
    for x in range(7):
        newDir = directory_name + "/" + str(x)
        createDir(newDir)

createDir('train')
createDir('validation')
createDir('test')

createSubDir("train")
createSubDir("validation")
createSubDir("test")
```

```python
# Reference of how to iterate over DataFrame rows
# Resource: https://www.geeksforgeeks.org/different-ways-to-iterate-over-rows-in-pandas-dataframe/
def addImagesToSet(dataset_name, df):
    for index, row in df.iterrows():
        os.chdir('/user/HS401/ob00564/Documents/COM3001/CK+')
        # label: Refers to folder number
        label = row['emotion']
        # img: Refers to the numPy array
        img = row['pixels']
        # data: Refers to PIL image
        # Resouce: https://stackoverflow.com/questions/55319949/pil-typeerror-cannot-handle-this-data-type
        data = Image.fromarray((img).astype(np.uint8))
        
        dirString = dataset_name + "/" + str(label)
        os.chdir(dirString)
        data.save(f'{label} Index {index}.png')

addImagesToSet(dataset_name='train', df = train_df)
addImagesToSet(dataset_name='validation', df = val_df)
addImagesToSet(dataset_name='test', df = test_df)
    
```
